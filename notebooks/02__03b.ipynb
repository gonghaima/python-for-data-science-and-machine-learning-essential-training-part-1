```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

```python
numbers_df = DataFrame(np.arange(0,90, 3).reshape(10, 3), index=['row 1', 'row 2', 'row 3', 'row 4', 'row 5', 'row 6', 'row 7', 'row 8', 'row 9', 'row 10'],columns=['col 1', 'col 2', 'col 3'])
```

```python
numbers_df
```

```python
numbers_df.iloc[0,1]
```

```python
numbers_df.iloc[0,1] = 20
```

```python
numbers_df
```

#### Comparison operators (> < = <= => == !=) and Masking.

```python
mask = numbers_df > 30
```

```python
mask
```

```python
numbers_df[mask]
```

```python
numbers_df[numbers_df > 30] = 0
numbers_df
```

```python
numbers_df.iloc[2:6, 1:3]
```
